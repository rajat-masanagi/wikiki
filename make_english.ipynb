```python
import re

# Define the regex pattern to match the desired characters
pattern = re.compile(r'[ !"#$%&\'()*+,\-./0-9:;<=>?@A-Za-z[\]^_`{|}~]+')

# Open the input and output files
input_file_path = 'wikipedia.txt'
output_file_path = 'wikipedia_eng.txt'

with open(input_file_path, 'r', encoding='utf-8') as input_file, \
     open(output_file_path, 'w', encoding='utf-8') as output_file:

    # Iterate through each line in the input file
    for line in input_file:
        # Use regex to filter out unwanted characters
        filtered_line = ''.join(pattern.findall(line))
        
        # Write the filtered line to the output file
        output_file.write(filtered_line + '\n')

print("Filtering complete.")
```
